```python
# Задание 2
# По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
# категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.
```

```python
consumption = pd.read_csv('power.csv')
consumption.head()
```

```python
Baltic = consumption[(consumption['country'] == 'Latvia') | (consumption['country'] == 'Lithuania') | (consumption['country'] == 'Estonia')]
#Baltic.head()
Baltic_4_12_21 = Baltic[(Baltic['category'] == 4) | (Baltic['category'] == 12) | (Baltic['category'] == 21)]
#Baltic_4_12_21
Baltic_years = Baltic_4_12_21[(Baltic_4_12_21['year'] <= 2010) & (Baltic_4_12_21['year'] >= 2005)]
sum(Baltic_years[Baltic_years['quantity'] > 0]['quantity'])
```
