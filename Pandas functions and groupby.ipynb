```python
import pandas as pd
```

```python
movies = pd.read_csv ('ratings.csv')
movies.head()
```

```python
def acceptance(rating):
    if rating <= 2:
        return ('отстой')
    elif rating <= 4:
        return('норм')
    return ('огонь')
```

```python
movies['class'] = movies['rating'].apply(acceptance)
movies.head()
```

```python
# Используем файл keywords.csv.
# Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую
# принадлежность определенному региону. Т. е. если поисковый запрос содержит название города
# региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит
# названия города, то ставим ‘undefined’.
# Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
# Результат классификации запишите в отдельный столбец region.
```

```python
keywords = pd.read_csv('keywords.csv')
keywords.head()
```

```python
def region(request):
    for FO, towns in geo_data.items():
        for town in towns:
            if town in request.split():
                return(FO)
    return('undefined')
keywords['region'] = keywords['keyword'].apply(region)
keywords[keywords['region'] != 'undefined']
#keywords.head()
```
