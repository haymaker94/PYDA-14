```python
# 1. перечень всех документов
documents = [
{'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
{'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
{'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
#2.перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)
directories = {
'1': ['2207 876234', '11-2'],
'2': ['10006'],
'3': []
}
command=''
def p_command():
    docnum = input('Введите номер документа, чтобы узнать имя юзера: ')
    for document in documents:
        if docnum == document['number']:
            print('Владелец документа: ', document['name'])
            return()
    else:
        print('Документ не найден в базе')

def s_command():
    docnum = input('Введите номер документа, чтобы узнать, где он хранится: ')
    for storage in directories:
        if docnum in list(directories[storage]):
            print('Документ лежит на полке №', storage)
            return()
    else:
        print('нет ни на одной полке')

def l_command():
    print('')
    print('Текущий список документов')
    for document in documents:
        docnum = document['number']
        doctype = document['type']
        username = document['name']
        for storage in directories:
            if docnum in directories[storage]:
                shelf = storage
        print(f'№ {docnum}, тип: {doctype}, юзер: {username}, лежит на полке № {shelf}')
    print('')

def as_command():
    newshelf = input('Введите желаемый номер для новой полки ')
    if not newshelf in directories:
        directories[newshelf] = []
    else:
        print('Такая полка уже имеется')
    print (f'Текущий список папок: {",".join(list(directories.keys()))}')

def ds_command():
    shelfToDel = input('Введите номер полки для удаления ')
    if shelfToDel in directories:
        if directories[shelfToDel] == []:
            del directories[shelfToDel]
        else:
            print('На полке есть документы. Сначала переместите документы на другую полку командой m')
    else:
        print('Такой полки нет')
    print (f'Текущий список папок: {",".join(list(directories.keys()))}')

def ad_command():
    docnum = input('Введите номер документа: ')
    doctype= input('Введите тип документа: ')
    username = input('Введите имя владельца: ')
    storage = input('Введите номер полки')
    if storage in directories:
        documents.append({'type': doctype, 'number': docnum, 'name': username})
        directories[storage].append(docnum)
    else:
        print('Такой полки нет. Сначала добавьте полку командой "as"')
    l_command()

def d_command():
    docnum = input('Введите номер документа: ')
    for docToDel in range(len(documents)):
        if documents[docToDel]['number'] == docnum:
            del documents[docToDel]
            for storage in directories:
                if docnum in directories[storage]:
                    directories[storage].remove(docnum)
            break
        else:
            print ('Документ не найден в базе')
            break
            
    l_command()
        

def m_command():
    docnum = input('Введите номер документа: ')
    doclist = [document['number'] for document in documents]
    if docnum in doclist:
        storage = input('Введите номер полки: ')
    else:
        print ('Документ не найден')
        l_command()
        return()

    storagelist = [shelf for shelf in directories]
    if storage in storagelist:
        for shelf in directories:
            for doc in directories[shelf]:
                if doc == docnum:
                    directories[shelf].remove(doc)
                    directories[storage].append (docnum)
                    break
    else:
        print(f'Такой полки нет, текущий перечень полок: {storagelist}')
    l_command()
        
                

command=''
while command != 'q':
    command = input('Введите команду: ')
    if command == 'p':
        p_command()
    elif command == 's':
        s_command()
    elif command == 'l':
        l_command()
    elif command == 'as':
        as_command()
    elif command == 'ds':
        ds_command()
    elif command == 'ad':
        ad_command()
    elif command == 'd':
        d_command()
    elif command == 'm':
        m_command()
```
